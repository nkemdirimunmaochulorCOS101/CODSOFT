```python
import tkinter as tk
from tkinter import messagebox

def calculate():
    try:    
        firstno = float(firstno_entry.get())
        secondno = float(secondno_entry.get())
    except ValueError:
        messagebox.showerror("Input Error", "Enter numbers")
        return

    operation = operation_entry.get().lower()

    if operation.lower() == "addition" or operation == "+":
        calc = firstno + secondno
        messagebox.showinfo("Addition", f"The addition of the first and second number is, {calc}.")
    
    elif operation.lower() == "subtraction" or operation == "-":
        calc = firstno - secondno
        messagebox.showinfo("Subtraction", f"The subtraction of the first and second number is, {calc}.")

    elif operation.lower() == "multiplication" or operation == "*":
        calc = firstno * secondno
        messagebox.showinfo("Multiplication", f"The multiplication of the first and second number is, {calc}.")
        
    elif operation.lower() == "division" or operation == "/":
        calc = firstno / secondno
        messagebox.showinfo("Division", f"The division of the first and second number is, {calc}.")
        
    else:
        messagebox.showerror("Error", "Operation not valid")
    

# Create main window
root = tk.Tk()
root.title("Calculator")
root.geometry("570x300")
root.configure(bg="white")

# Create username label and entry
firstno_label = tk.Label(root, text="Enter First Number:")
firstno_label.pack()
firstno_entry = tk.Entry(root)
firstno_entry.pack()

# Create department label and entry
secondno_label = tk.Label(root, text="Enter Second Number:")
secondno_label.pack()
secondno_entry = tk.Entry(root)
secondno_entry.pack()

# Create department label and entry
operation_label = tk.Label(root, text="Enter Operation of choice:")
operation_label.pack()
operation_entry = tk.Entry(root)
operation_entry.pack()

# Create submit button
calculate_button = tk.Button(root, text="Calculate", command=calculate)
calculate_button.pack()

# Run the main event loop
root.mainloop()
```
